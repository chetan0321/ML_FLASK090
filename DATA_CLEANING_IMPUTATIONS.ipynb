```python
# prompt: simple dataframe with missing values for helmet detection use case data is age height helmet  status gender experince
# replacing missing values with mean ,median and mode
# show the result for all imputations'

import pandas as pd
import numpy as np

# Create a sample DataFrame with missing values
data = {'age': [25, 30, np.nan, 40, 28],
        'height': [170, 175, 165, np.nan, 180],
        'helmet_status': [1, 0, 1, 0, np.nan],  # 1 for wearing, 0 for not wearing
        'gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
        'experience': [2, 5, np.nan, 3, 7]}
df = pd.DataFrame(data)

# Mean imputation for numerical columns
df_mean = df.copy()
numerical_cols = ['age', 'height']
for col in numerical_cols:
    df_mean[col].fillna(df_mean[col].mean(), inplace=True)

# Median imputation for numerical columns
df_median = df.copy()
for col in numerical_cols:
    df_median[col].fillna(df_median[col].median(), inplace=True)

# Mode imputation for categorical columns
df_mode = df.copy()
categorical_cols = ['helmet_status', 'gender', 'experience']

for col in categorical_cols:
  if pd.api.types.is_numeric_dtype(df_mode[col]):
    df_mode[col].fillna(df_mode[col].mode()[0], inplace=True)
  else:
    df_mode[col].fillna(df_mode[col].mode()[0], inplace=True)


print("Original DataFrame:")
print(df)

print("\nDataFrame with mean imputation:")
print(df_mean)

print("\nDataFrame with median imputation:")
print(df_median)

print("\nDataFrame with mode imputation:")
df_mode
```

```python
# prompt: simple dataframe with missing values for helmet detection use case data is age height helmet status gender experience
# replacing missing values with mean, median, and mode
# show the result for all imputations'

import pandas as pd
import numpy as np

# Create a sample DataFrame with missing values
data = {'age': [25, 30, np.nan, 40, 28],
        'height': [170, 175, 165, np.nan, 180],
        'helmet_status': [1, 0, 1, 0, np.nan],  # 1 for wearing, 0 for not wearing
        'gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
        'experience': [2, 5, np.nan, 3, 7]}
df = pd.DataFrame(data)

# Mean imputation for numerical columns
df_mean = df.copy()
numerical_cols = ['age', 'height']
df_mean.fillna({col: df_mean[col].mean() for col in numerical_cols}, inplace=True)  # Using df.fillna() directly


# Median imputation for numerical columns
df_median = df.copy()
df_median.fillna({col: df_median[col].median() for col in numerical_cols}, inplace=True)  # Using df.fillna() directly


# Mode imputation for categorical columns
df_mode = df.copy()
categorical_cols = ['helmet_status', 'gender', 'experience']
df_mode.fillna({col: df_mode[col].mode()[0] for col in categorical_cols}, inplace=True)  # Using df.fillna() directly



print("Original DataFrame:")
print(df)

print("\nDataFrame with mean imputation:")
print(df_mean)

print("\nDataFrame with median imputation:")
print(df_median)

print("\nDataFrame with mode imputation:")
df_mode
```
