```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset with outliers
data = {'values': [10, 12, 15, 18, 20, 22, 25, 100, 110, 15, 17, 19, 21, 23, 27]}  # Added some outliers
df = pd.DataFrame(data)

# IQR calculation
Q1 = df['values'].quantile(0.25)
Q3 = df['values'].quantile(0.75)
IQR = Q3 - Q1

# Outlier thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['values'] < lower_bound) | (df['values'] > upper_bound)]

# Print outlier thresholds and identified outliers
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print("\nOutliers:")
print(outliers)

# Box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='values', data=df)
plt.title('Box Plot of Values with Outliers')
plt.show()


# Violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(x='values', data=df, inner="box") # Added inner="box"
plt.title('Violin Plot of Values with Outliers and Box Inside')
plt.show()
```
