```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Sample DataFrame (replace with your actual data)
data = {'col1': [1, 2, 3, 4, 5],
        'col2': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Normalize the dataset
normalized_data = scaler.fit_transform(df)

# Create a new DataFrame with normalized values
normalized_df = pd.DataFrame(normalized_data, columns=df.columns)

# Print the normalized DataFrame
normalized_df
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Sample DataFrame (replace with your actual data)
data = {'col1': [1, 2, 3, 4, 5],
        'col2': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

# Initialize StandardScaler
scaler = StandardScaler()

# Standardize the dataset
standerized_data = scaler.fit_transform(df)

# Create a new DataFrame with standardized values
standerized_df = pd.DataFrame(standerized_data, columns=df.columns)

# Print the standardized DataFrame
print(standerized_df)

# Check the mean and standard deviation of the standardized DataFrame
print("\nMean of standardized DataFrame:")
print(standerized_df.mean())

print("\nStandard Deviation of standardized DataFrame:")
print(standerized_df.std())
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Sample DataFrame (replace with your actual data)
data = {'col1': ['A', 'B', 'A', 'C', 'B'],
        'col2': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

# Initialize LabelEncoder
encoder = LabelEncoder()

df['col1_encoded'] = encoder.fit_transform(df['col1'])

df
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Create a sample DataFrame
data = {'customer_id': [1, 2, 3, 4, 5],
        'gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
        'subscription': ['Basic', 'Premium', 'Basic', 'Premium', 'Free']}
df = pd.DataFrame(data)

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply label encoding to the 'gender' column
df['gender_encoded'] = label_encoder.fit_transform(df['gender'])

# Apply label encoding to the 'subscription' column
df['subscription_encoded'] = label_encoder.fit_transform(df['subscription'])

# Print the DataFrame with encoded columns
df
```

```python
# prompt: create a data and convert the data into dataframe and initialize one hot encoder
# select coulms toencode
# fit and transform the data

import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Create a sample DataFrame
data = {'color': ['red', 'green', 'blue', 'red', 'green'],
        'size': ['small', 'medium', 'large', 'small', 'medium'],
        'price': [10, 20, 30, 10, 20]}
df = pd.DataFrame(data)

# Initialize OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore') # sparse=False for dense output

# Select columns to encode
columns_to_encode = ['color', 'size']

# Fit and transform the selected columns
encoded_data = encoder.fit_transform(df[columns_to_encode])

# Create a new DataFrame with the encoded data
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(columns_to_encode))

# Concatenate the encoded DataFrame with the original DataFrame
df = pd.concat([df, encoded_df], axis=1)
df
```
